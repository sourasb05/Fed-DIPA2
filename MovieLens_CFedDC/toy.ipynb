```python
import gzip
import json
import pandas as pd
import urllib.request

# URL for Books category (Amazon 2018 v2)
url = "https://datarepo.eng.ucsd.edu/mcauley_group/data/amazon_v2/categoryFiles/Books.json.gz"

# Download the file
urllib.request.urlretrieve(url, "Books.json.gz")

# Read JSON lines
data = []
with gzip.open("Books.json.gz", 'rt') as f:
    for line in f:
        try:
            review = json.loads(line)
            data.append({
                'reviewerID': review.get('reviewerID'),
                'reviewText': review.get('reviewText'),
                'overall': review.get('overall'),
                'helpfulVotes': review.get('helpful', [0, 0])[0]  # helpful[0] is yes-votes
            })
        except json.JSONDecodeError:
            continue

df = pd.DataFrame(data)
print(df.head())
```
