```python
import statistics

# assume you already have these four lists in your namespace:
# min_rf_1, min_rf_2, min_rf_3, min_rf_4

all_lists = [min_rf_1, min_rf_2, min_rf_3, min_rf_4]
labels    = ['delta=0.1', 'delta=0.5', 'delta=0.8', 'delta=1.0']

for label, data in zip(labels, all_lists):
    if not data:
        print(f"{label}: no values")
        continue

    meu = statistics.mean(data)
    sigma = statistics.stdev(data)     # sample standard deviation (ddof=1)
    # _pop = statistics.pstdev(data)  # if you want populationstd instead

    print(f"{label:8s}: {meu:.3f}  (std = {sigma:.3f})")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import statistics

all_lists = [min_rf_1, min_rf_2, min_rf_3, min_rf_4]
deltas   = np.array([0.1, 0.5, 0.8, 1.0])

# Compute means and standard deviations
means = np.array([statistics.mean(lst) for lst in all_lists])
stds  = np.array([statistics.stdev(lst) for lst in all_lists])

# Fit a linear trendline
coef  = np.polyfit(deltas, means, deg=1)
trend = np.poly1d(coef)

# Generate smooth x for the trendline
x_fit = np.linspace(deltas.min(), deltas.max(), 100)
y_fit = trend(x_fit)

plt.figure()

# Plot 1 std shadow around the trendline
upper = trend(deltas) + stds
lower = trend(deltas) - stds
plt.fill_between(deltas, lower, upper,
                 color='gray', alpha=0.3,
                 label='1 Std Shadow')

# Plot the dashed trendline
plt.plot(x_fit, y_fit, linestyle='--', color='black',
         label=f'Trend: y={coef[0]:.2f}x+{coef[1]:.2f}')

# Plot mean points with error bars
plt.errorbar(deltas, means, yerr=stds,
             marker='o', linestyle='',
             capsize=5, color='orange',
             label='Mean  Std')

# Add horizontal/vertical margins so labels don't get clipped
plt.margins(x=0.05, y=0.1)

# Annotate each mean value; shift last one left
for i, (x, m) in enumerate(zip(deltas, means)):
    if i == len(deltas) - 1:
        dx, ha = -5, 'right'
    else:
        dx, ha = 5, 'left'

    plt.annotate(
        f"{m:.2f}",
        xy=(x, m),
        xytext=(dx, 5),          # offset in points
        textcoords="offset points",
        ha=ha,                   # horizontal alignment
        va="bottom",
        fontsize=9,
        clip_on=False            # allow text outside axes
    )

plt.xlabel(r'$\delta$', fontsize=16)
plt.ylabel('Mean of Min(info_cmae)', fontsize=16)
# plt.title('Mean  1 Std Dev with Trendline Shadow')
plt.xticks(deltas)
# plt.legend()
plt.tight_layout()
plt.savefig('kappa_constant_delta_min_info_cmae.png', dpi=300)
plt.show()
```
