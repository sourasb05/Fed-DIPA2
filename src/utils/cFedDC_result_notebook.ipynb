```python
import os, glob, json, statistics
import numpy as np

# ------------------------------
# Config
# ------------------------------
EXP_COUNT = 5
BASE_DIR = "/proj/sourasb-220503/TCSS-DIPA2/Fed-DIPA2/results/client_level/CFedDC_KT_RL_rl1_C2_ViT-L_14_336px"
EXP_TEMPLATE = "exp_no_{i}_CFedDC_KT_RL_lambda_min_0.4_lambda_max_0.6_kappa_1.0_delta_1.0_30"
SUBDIR = "local_test"  # or "local_val"

# ------------------------------
# Helpers
# ------------------------------
def safe_mean_std(values):
    vals = [v for v in values if v is not None]
    if not vals: return (None, None)
    if len(vals) == 1: return (vals[0], 0.0)
    return (statistics.mean(vals), statistics.stdev(vals))

def mean_or_none(x): return float(np.mean(x)) if len(x) > 0 else None
def fmt(x): return f"{x:.3f}" if x is not None else "NA"

def accumulate_triplet(accI, accSO, accSOW, triplet):
    if triplet is None: return
    i, so, sow = triplet
    accI.append(i); accSO.append(so); accSOW.append(sow)

def extract_user_metrics_extended(file_path):
    """
    Returns:
      category: "Resourceful user" | "Resourceless user" | None
      min_cmae: float or None
      f1_at_min: (I, SO, SOW) or None
      f1_max_triplet: (max F1I, max F1SO, max F1SOW) or None  # NEW
    """
    try:
        with open(file_path, "r") as f:
            d = json.load(f)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        return None, None, None, None

    category = d.get("resource_category")
    info_cmae = d.get("validation_results", {}).get("info_cmae", [])
    f1 = d.get("validation_results", {}).get("F1Score", [])

    if not f1 or not isinstance(f1, list):
        return category, None, None, None

    # --- F1@minCMAE ---
    min_cmae, f1_at_min = None, None
    if info_cmae and isinstance(info_cmae, list):
        try:
            min_cmae = float(min(info_cmae))
            min_idx  = info_cmae.index(min_cmae)
            if 0 <= min_idx < len(f1):
                row = f1[min_idx]
                if isinstance(row, (list, tuple)) and len(row) >= 3:
                    f1_at_min = (float(row[0]), float(row[1]), float(row[2]))
        except Exception:
            pass

    # --- MaxF1 (irrespective of CMAE) ---
    # f1 is a list of [I, SO, SOW] per step; take column-wise max
    try:
        cols = list(zip(*f1))  # [(all I), (all SO), (all SOW)]
        if len(cols) >= 3:
            f1_max_triplet = (float(max(cols[0])), float(max(cols[1])), float(max(cols[2])))
        else:
            f1_max_triplet = None
    except Exception:
        f1_max_triplet = None

    return category, min_cmae, f1_at_min, f1_max_triplet

# ------------------------------
# Overall accumulators (experiment-level means)
# ------------------------------
all_cmae_all, all_cmae_rf, all_cmae_rl = [], [], []

# F1@minCMAE (means per experiment)
all_f1min_I_all, all_f1min_SO_all, all_f1min_SOW_all = [], [], []
all_f1min_I_rf,  all_f1min_SO_rf,  all_f1min_SOW_rf  = [], [], []
all_f1min_I_rl,  all_f1min_SO_rl,  all_f1min_SOW_rl  = [], [], []

# MaxF1 (means per experiment)  # NEW
all_f1max_I_all, all_f1max_SO_all, all_f1max_SOW_all = [], [], []
all_f1max_I_rf,  all_f1max_SO_rf,  all_f1max_SOW_rf  = [], [], []
all_f1max_I_rl,  all_f1max_SO_rl,  all_f1max_SOW_rl  = [], [], []

# ------------------------------
# Main
# ------------------------------
for i in range(EXP_COUNT):
    exp_dir   = os.path.join(BASE_DIR, EXP_TEMPLATE.format(i=i), SUBDIR)
    file_list = glob.glob(os.path.join(exp_dir, "*.json"))
    if not file_list:
        print(f"[WARN] No JSON files in {exp_dir}")
        continue

    # Per-experiment buckets
    cmae_all, cmae_rf, cmae_rl = [], [], []
    f1min_I_all, f1min_SO_all, f1min_SOW_all = [], [], []
    f1min_I_rf,  f1min_SO_rf,  f1min_SOW_rf  = [], [], []
    f1min_I_rl,  f1min_SO_rl,  f1min_SOW_rl  = [], [], []

    f1max_I_all, f1max_SO_all, f1max_SOW_all = [], [], []
    f1max_I_rf,  f1max_SO_rf,  f1max_SOW_rf  = [], [], []
    f1max_I_rl,  f1max_SO_rl,  f1max_SOW_rl  = [], [], []

    print(f"\n=== Experiment {i} ===")
    # Iterate users
    for fp in file_list:
        category, min_cmae, f1_at_min, f1_max = extract_user_metrics_extended(fp)

        # CMAE
        if min_cmae is not None:
            cmae_all.append(min_cmae)
            if category == "Resourceful user":   cmae_rf.append(min_cmae)
            elif category == "Resourceless user": cmae_rl.append(min_cmae)

        # F1@minCMAE
        accumulate_triplet(f1min_I_all, f1min_SO_all, f1min_SOW_all, f1_at_min)
        if category == "Resourceful user":
            accumulate_triplet(f1min_I_rf, f1min_SO_rf, f1min_SOW_rf, f1_at_min)
        elif category == "Resourceless user":
            accumulate_triplet(f1min_I_rl, f1min_SO_rl, f1min_SOW_rl, f1_at_min)

        # MaxF1 (irrespective of CMAE)  # NEW
        accumulate_triplet(f1max_I_all, f1max_SO_all, f1max_SOW_all, f1_max)
        if category == "Resourceful user":
            accumulate_triplet(f1max_I_rf, f1max_SO_rf, f1max_SOW_rf, f1_max)
        elif category == "Resourceless user":
            accumulate_triplet(f1max_I_rl, f1max_SO_rl, f1max_SOW_rl, f1_max)

    # Per-experiment means (CMAE)
    m_all, s_all = safe_mean_std(cmae_all)
    m_rf,  s_rf  = safe_mean_std(cmae_rf)
    m_rl,  s_rl  = safe_mean_std(cmae_rl)

    if m_all is not None: all_cmae_all.append(m_all)
    if m_rf  is not None: all_cmae_rf.append(m_rf)
    if m_rl  is not None: all_cmae_rl.append(m_rl)

    # Per-experiment means (F1@minCMAE)
    f1min_all_I, f1min_all_SO, f1min_all_SOW = mean_or_none(f1min_I_all), mean_or_none(f1min_SO_all), mean_or_none(f1min_SOW_all)
    f1min_rf_I,  f1min_rf_SO,  f1min_rf_SOW  = mean_or_none(f1min_I_rf),  mean_or_none(f1min_SO_rf),  mean_or_none(f1min_SOW_rf)
    f1min_rl_I,  f1min_rl_SO,  f1min_rl_SOW  = mean_or_none(f1min_I_rl),  mean_or_none(f1min_SO_rl),  mean_or_none(f1min_SOW_rl)

    # Per-experiment means (MaxF1)  # NEW
    f1max_all_I, f1max_all_SO, f1max_all_SOW = mean_or_none(f1max_I_all), mean_or_none(f1max_SO_all), mean_or_none(f1max_SOW_all)
    f1max_rf_I,  f1max_rf_SO,  f1max_rf_SOW  = mean_or_none(f1max_I_rf),  mean_or_none(f1max_SO_rf),  mean_or_none(f1max_SOW_rf)
    f1max_rl_I,  f1max_rl_SO,  f1max_rl_SOW  = mean_or_none(f1max_I_rl),  mean_or_none(f1max_SO_rl),  mean_or_none(f1max_SOW_rl)

    # Store for overall (as experiment-level means)
    for val, buf in [
        (f1min_all_I, all_f1min_I_all), (f1min_all_SO, all_f1min_SO_all), (f1min_all_SOW, all_f1min_SOW_all),
        (f1min_rf_I,  all_f1min_I_rf),  (f1min_rf_SO,  all_f1min_SO_rf),  (f1min_rf_SOW,  all_f1min_SOW_rf),
        (f1min_rl_I,  all_f1min_I_rl),  (f1min_rl_SO,  all_f1min_SO_rl),  (f1min_rl_SOW,  all_f1min_SOW_rl),
        (f1max_all_I, all_f1max_I_all), (f1max_all_SO, all_f1max_SO_all), (f1max_all_SOW, all_f1max_SOW_all),
        (f1max_rf_I,  all_f1max_I_rf),  (f1max_rf_SO,  all_f1max_SO_rf),  (f1max_rf_SOW,  all_f1max_SOW_rf),
        (f1max_rl_I,  all_f1max_I_rl),  (f1max_rl_SO,  all_f1max_SO_rl),  (f1max_rl_SOW,  all_f1max_SOW_rl),
    ]:
        if val is not None: buf.append(val)

    # --- Pretty per-experiment prints ---
    print(f"exp_{i}:::Average minimum info_cmae  All:{fmt(m_all)}  RF:{fmt(m_rf)}  RL:{fmt(m_rl)}")
    print(f"exp_{i}:::F1@minCMAE (All)        I:{fmt(f1min_all_I)}  SO:{fmt(f1min_all_SO)}  SOW:{fmt(f1min_all_SOW)}")
    print(f"exp_{i}:::F1@minCMAE (Resourceful) I:{fmt(f1min_rf_I)}   SO:{fmt(f1min_rf_SO)}   SOW:{fmt(f1min_rf_SOW)}")
    print(f"exp_{i}:::F1@minCMAE (Resourceless)I:{fmt(f1min_rl_I)}   SO:{fmt(f1min_rl_SO)}   SOW:{fmt(f1min_rl_SOW)}")
    print(f"exp_{i}:::MaxF1 (All)              I:{fmt(f1max_all_I)}  SO:{fmt(f1max_all_SO)}  SOW:{fmt(f1max_all_SOW)}")
    print(f"exp_{i}:::MaxF1 (Resourceful)      I:{fmt(f1max_rf_I)}   SO:{fmt(f1max_rf_SO)}   SOW:{fmt(f1max_rf_SOW)}")
    print(f"exp_{i}:::MaxF1 (Resourceless)     I:{fmt(f1max_rl_I)}   SO:{fmt(f1max_rl_SO)}   SOW:{fmt(f1max_rl_SOW)}")

# ------------------------------
# Overall summaries (mean  stdev across experiments)
# ------------------------------
def print_overall(title, values):
    m, s = safe_mean_std(values)
    print(f"{title}: {fmt(m)}/{fmt(s)}")

print("\n================ OVERALL SUMMARY ================")
# CMAE
print_overall("Overall:::Average minimum info_cmae (All)", all_cmae_all)
print_overall("Overall:::Average minimum info_cmae (Resourceful)", all_cmae_rf)
print_overall("Overall:::Average minimum info_cmae (Resourceless)", all_cmae_rl)

# F1@minCMAE
print_overall("Overall:::F1@minCMAE (All)  Information types", all_f1min_I_all)
print_overall("Overall:::F1@minCMAE (All)  Sharing Owner",   all_f1min_SO_all)
print_overall("Overall:::F1@minCMAE (All)  Sharing Other",   all_f1min_SOW_all)

print_overall("Overall:::F1@minCMAE (Resourceful)  Information types", all_f1min_I_rf)
print_overall("Overall:::F1@minCMAE (Resourceful)  Sharing Owner",   all_f1min_SO_rf)
print_overall("Overall:::F1@minCMAE (Resourceful)  Sharing Other",   all_f1min_SOW_rf)

print_overall("Overall:::F1@minCMAE (Resourceless)  Information types", all_f1min_I_rl)
print_overall("Overall:::F1@minCMAE (Resourceless)  Sharing Owner",   all_f1min_SO_rl)
print_overall("Overall:::F1@minCMAE (Resourceless)  Sharing Other",   all_f1min_SOW_rl)

# MaxF1 (irrespective of CMAE)  # NEW
print_overall("Overall:::MaxF1 (All)  Information types", all_f1max_I_all)
print_overall("Overall:::MaxF1 (All)  Sharing Owner",     all_f1max_SO_all)
print_overall("Overall:::MaxF1 (All)  Sharing Other",     all_f1max_SOW_all)

print_overall("Overall:::MaxF1 (Resourceful)  Information types", all_f1max_I_rf)
print_overall("Overall:::MaxF1 (Resourceful)  Sharing Owner",     all_f1max_SO_rf)
print_overall("Overall:::MaxF1 (Resourceful)  Sharing Other",     all_f1max_SOW_rf)

print_overall("Overall:::MaxF1 (Resourceless)  Information types", all_f1max_I_rl)
print_overall("Overall:::MaxF1 (Resourceless)  Sharing Owner",     all_f1max_SO_rl)
print_overall("Overall:::MaxF1 (Resourceless)  Sharing Other",     all_f1max_SOW_rl)
```

```python
import os, glob, json, statistics
import numpy as np

# ------------------------------
# Config
# ------------------------------
EXP_COUNT = 5
BASE_DIR = "/proj/sourasb-220503/TCSS-DIPA2/Fed-DIPA2/results/client_level/CFedDC_KT_RL_rl1_C2_timm_mobilevit"
EXP_TEMPLATE = "exp_no_{i}_CFedDC_KT_RL_lambda_min_0.4_lambda_max_0.6_kappa_1.0_delta_1.0_30"
SUBDIR = "local_test"  # or "local_val"

# ------------------------------
# Helpers
# ------------------------------
def safe_mean_std(values):
    vals = [v for v in values if v is not None]
    if not vals: return (None, None)
    if len(vals) == 1: return (vals[0], 0.0)
    return (statistics.mean(vals), statistics.stdev(vals))

def mean_or_none(x): return float(np.mean(x)) if len(x) > 0 else None
def fmt(x): return f"{x:.3f}" if x is not None else "NA"

def accumulate_triplet(accI, accSO, accSOW, triplet):
    if triplet is None: return
    i, so, sow = triplet
    accI.append(i); accSO.append(so); accSOW.append(sow)

def extract_user_metrics_extended(file_path):
    """
    Returns:
      category: "Resourceful user" | "Resourceless user" | None
      min_cmae: float or None
      f1_at_min: (I, SO, SOW) or None
      f1_max_triplet: (max F1I, max F1SO, max F1SOW) or None  # NEW
    """
    try:
        with open(file_path, "r") as f:
            d = json.load(f)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        return None, None, None, None

    category = d.get("resource_category")
    info_cmae = d.get("validation_results", {}).get("info_cmae", [])
    f1 = d.get("validation_results", {}).get("F1Score", [])

    if not f1 or not isinstance(f1, list):
        return category, None, None, None

    # --- F1@minCMAE ---
    min_cmae, f1_at_min = None, None
    if info_cmae and isinstance(info_cmae, list):
        try:
            min_cmae = float(min(info_cmae))
            min_idx  = info_cmae.index(min_cmae)
            if 0 <= min_idx < len(f1):
                row = f1[min_idx]
                if isinstance(row, (list, tuple)) and len(row) >= 3:
                    f1_at_min = (float(row[0]), float(row[1]), float(row[2]))
        except Exception:
            pass

    # --- MaxF1 (irrespective of CMAE) ---
    # f1 is a list of [I, SO, SOW] per step; take column-wise max
    try:
        cols = list(zip(*f1))  # [(all I), (all SO), (all SOW)]
        if len(cols) >= 3:
            f1_max_triplet = (float(max(cols[0])), float(max(cols[1])), float(max(cols[2])))
        else:
            f1_max_triplet = None
    except Exception:
        f1_max_triplet = None

    return category, min_cmae, f1_at_min, f1_max_triplet

# ------------------------------
# Overall accumulators (experiment-level means)
# ------------------------------
all_cmae_all, all_cmae_rf, all_cmae_rl = [], [], []

# F1@minCMAE (means per experiment)
all_f1min_I_all, all_f1min_SO_all, all_f1min_SOW_all = [], [], []
all_f1min_I_rf,  all_f1min_SO_rf,  all_f1min_SOW_rf  = [], [], []
all_f1min_I_rl,  all_f1min_SO_rl,  all_f1min_SOW_rl  = [], [], []

# MaxF1 (means per experiment)  # NEW
all_f1max_I_all, all_f1max_SO_all, all_f1max_SOW_all = [], [], []
all_f1max_I_rf,  all_f1max_SO_rf,  all_f1max_SOW_rf  = [], [], []
all_f1max_I_rl,  all_f1max_SO_rl,  all_f1max_SOW_rl  = [], [], []

# ------------------------------
# Main
# ------------------------------
for i in range(EXP_COUNT):
    exp_dir   = os.path.join(BASE_DIR, EXP_TEMPLATE.format(i=i), SUBDIR)
    file_list = glob.glob(os.path.join(exp_dir, "*.json"))
    if not file_list:
        print(f"[WARN] No JSON files in {exp_dir}")
        continue

    # Per-experiment buckets
    cmae_all, cmae_rf, cmae_rl = [], [], []
    f1min_I_all, f1min_SO_all, f1min_SOW_all = [], [], []
    f1min_I_rf,  f1min_SO_rf,  f1min_SOW_rf  = [], [], []
    f1min_I_rl,  f1min_SO_rl,  f1min_SOW_rl  = [], [], []

    f1max_I_all, f1max_SO_all, f1max_SOW_all = [], [], []
    f1max_I_rf,  f1max_SO_rf,  f1max_SOW_rf  = [], [], []
    f1max_I_rl,  f1max_SO_rl,  f1max_SOW_rl  = [], [], []

    print(f"\n=== Experiment {i} ===")
    # Iterate users
    for fp in file_list:
        category, min_cmae, f1_at_min, f1_max = extract_user_metrics_extended(fp)

        # CMAE
        if min_cmae is not None:
            cmae_all.append(min_cmae)
            if category == "Resourceful user":   cmae_rf.append(min_cmae)
            elif category == "Resourceless user": cmae_rl.append(min_cmae)

        # F1@minCMAE
        accumulate_triplet(f1min_I_all, f1min_SO_all, f1min_SOW_all, f1_at_min)
        if category == "Resourceful user":
            accumulate_triplet(f1min_I_rf, f1min_SO_rf, f1min_SOW_rf, f1_at_min)
        elif category == "Resourceless user":
            accumulate_triplet(f1min_I_rl, f1min_SO_rl, f1min_SOW_rl, f1_at_min)

        # MaxF1 (irrespective of CMAE)  # NEW
        accumulate_triplet(f1max_I_all, f1max_SO_all, f1max_SOW_all, f1_max)
        if category == "Resourceful user":
            accumulate_triplet(f1max_I_rf, f1max_SO_rf, f1max_SOW_rf, f1_max)
        elif category == "Resourceless user":
            accumulate_triplet(f1max_I_rl, f1max_SO_rl, f1max_SOW_rl, f1_max)

    # Per-experiment means (CMAE)
    m_all, s_all = safe_mean_std(cmae_all)
    m_rf,  s_rf  = safe_mean_std(cmae_rf)
    m_rl,  s_rl  = safe_mean_std(cmae_rl)

    if m_all is not None: all_cmae_all.append(m_all)
    if m_rf  is not None: all_cmae_rf.append(m_rf)
    if m_rl  is not None: all_cmae_rl.append(m_rl)

    # Per-experiment means (F1@minCMAE)
    f1min_all_I, f1min_all_SO, f1min_all_SOW = mean_or_none(f1min_I_all), mean_or_none(f1min_SO_all), mean_or_none(f1min_SOW_all)
    f1min_rf_I,  f1min_rf_SO,  f1min_rf_SOW  = mean_or_none(f1min_I_rf),  mean_or_none(f1min_SO_rf),  mean_or_none(f1min_SOW_rf)
    f1min_rl_I,  f1min_rl_SO,  f1min_rl_SOW  = mean_or_none(f1min_I_rl),  mean_or_none(f1min_SO_rl),  mean_or_none(f1min_SOW_rl)

    # Per-experiment means (MaxF1)  # NEW
    f1max_all_I, f1max_all_SO, f1max_all_SOW = mean_or_none(f1max_I_all), mean_or_none(f1max_SO_all), mean_or_none(f1max_SOW_all)
    f1max_rf_I,  f1max_rf_SO,  f1max_rf_SOW  = mean_or_none(f1max_I_rf),  mean_or_none(f1max_SO_rf),  mean_or_none(f1max_SOW_rf)
    f1max_rl_I,  f1max_rl_SO,  f1max_rl_SOW  = mean_or_none(f1max_I_rl),  mean_or_none(f1max_SO_rl),  mean_or_none(f1max_SOW_rl)

    # Store for overall (as experiment-level means)
    for val, buf in [
        (f1min_all_I, all_f1min_I_all), (f1min_all_SO, all_f1min_SO_all), (f1min_all_SOW, all_f1min_SOW_all),
        (f1min_rf_I,  all_f1min_I_rf),  (f1min_rf_SO,  all_f1min_SO_rf),  (f1min_rf_SOW,  all_f1min_SOW_rf),
        (f1min_rl_I,  all_f1min_I_rl),  (f1min_rl_SO,  all_f1min_SO_rl),  (f1min_rl_SOW,  all_f1min_SOW_rl),
        (f1max_all_I, all_f1max_I_all), (f1max_all_SO, all_f1max_SO_all), (f1max_all_SOW, all_f1max_SOW_all),
        (f1max_rf_I,  all_f1max_I_rf),  (f1max_rf_SO,  all_f1max_SO_rf),  (f1max_rf_SOW,  all_f1max_SOW_rf),
        (f1max_rl_I,  all_f1max_I_rl),  (f1max_rl_SO,  all_f1max_SO_rl),  (f1max_rl_SOW,  all_f1max_SOW_rl),
    ]:
        if val is not None: buf.append(val)

    # --- Pretty per-experiment prints ---
    print(f"exp_{i}:::Average minimum info_cmae  All:{fmt(m_all)}  RF:{fmt(m_rf)}  RL:{fmt(m_rl)}")
    print(f"exp_{i}:::F1@minCMAE (All)        I:{fmt(f1min_all_I)}  SO:{fmt(f1min_all_SO)}  SOW:{fmt(f1min_all_SOW)}")
    print(f"exp_{i}:::F1@minCMAE (Resourceful) I:{fmt(f1min_rf_I)}   SO:{fmt(f1min_rf_SO)}   SOW:{fmt(f1min_rf_SOW)}")
    print(f"exp_{i}:::F1@minCMAE (Resourceless)I:{fmt(f1min_rl_I)}   SO:{fmt(f1min_rl_SO)}   SOW:{fmt(f1min_rl_SOW)}")
    print(f"exp_{i}:::MaxF1 (All)              I:{fmt(f1max_all_I)}  SO:{fmt(f1max_all_SO)}  SOW:{fmt(f1max_all_SOW)}")
    print(f"exp_{i}:::MaxF1 (Resourceful)      I:{fmt(f1max_rf_I)}   SO:{fmt(f1max_rf_SO)}   SOW:{fmt(f1max_rf_SOW)}")
    print(f"exp_{i}:::MaxF1 (Resourceless)     I:{fmt(f1max_rl_I)}   SO:{fmt(f1max_rl_SO)}   SOW:{fmt(f1max_rl_SOW)}")

# ------------------------------
# Overall summaries (mean  stdev across experiments)
# ------------------------------
def print_overall(title, values):
    m, s = safe_mean_std(values)
    print(f"{title}: {fmt(m)}/{fmt(s)}")

print("\n================ OVERALL SUMMARY ================")
# CMAE
print_overall("Overall:::Average minimum info_cmae (All)", all_cmae_all)
print_overall("Overall:::Average minimum info_cmae (Resourceful)", all_cmae_rf)
print_overall("Overall:::Average minimum info_cmae (Resourceless)", all_cmae_rl)

# F1@minCMAE
print_overall("Overall:::F1@minCMAE (All)  Information types", all_f1min_I_all)
print_overall("Overall:::F1@minCMAE (All)  Sharing Owner",   all_f1min_SO_all)
print_overall("Overall:::F1@minCMAE (All)  Sharing Other",   all_f1min_SOW_all)

print_overall("Overall:::F1@minCMAE (Resourceful)  Information types", all_f1min_I_rf)
print_overall("Overall:::F1@minCMAE (Resourceful)  Sharing Owner",   all_f1min_SO_rf)
print_overall("Overall:::F1@minCMAE (Resourceful)  Sharing Other",   all_f1min_SOW_rf)

print_overall("Overall:::F1@minCMAE (Resourceless)  Information types", all_f1min_I_rl)
print_overall("Overall:::F1@minCMAE (Resourceless)  Sharing Owner",   all_f1min_SO_rl)
print_overall("Overall:::F1@minCMAE (Resourceless)  Sharing Other",   all_f1min_SOW_rl)

# MaxF1 (irrespective of CMAE)  # NEW
print_overall("Overall:::MaxF1 (All)  Information types", all_f1max_I_all)
print_overall("Overall:::MaxF1 (All)  Sharing Owner",     all_f1max_SO_all)
print_overall("Overall:::MaxF1 (All)  Sharing Other",     all_f1max_SOW_all)

print_overall("Overall:::MaxF1 (Resourceful)  Information types", all_f1max_I_rf)
print_overall("Overall:::MaxF1 (Resourceful)  Sharing Owner",     all_f1max_SO_rf)
print_overall("Overall:::MaxF1 (Resourceful)  Sharing Other",     all_f1max_SOW_rf)

print_overall("Overall:::MaxF1 (Resourceless)  Information types", all_f1max_I_rl)
print_overall("Overall:::MaxF1 (Resourceless)  Sharing Owner",     all_f1max_SO_rl)
print_overall("Overall:::MaxF1 (Resourceless)  Sharing Other",     all_f1max_SOW_rl)
```
