```python
import pandas as pd
import glob
import os

# Set the folder path
folder_path = "/proj/sourasb-220503/TCSS-DIPA2/Fed-DIPA2/dataset_files/dynamic_FedDcprivacy"
# Get all CSV files
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Initialize storage
file_row_counts = {}
total_train_rows = 0
total_val_rows = 0
total_test_rows = 0

train_file_count = 0
val_file_count = 0
test_file_count = 0

# Process each file
for file in csv_files:
    try:
        df = pd.read_csv(file)
        num_rows = len(df)
        file_name = os.path.basename(file)
        file_row_counts[file_name] = num_rows

        # Aggregation by prefix
        if file_name.startswith("train"):
            total_train_rows += num_rows
            train_file_count += 1
        elif file_name.startswith("val"):
            total_val_rows += num_rows
            val_file_count += 1
        elif file_name.startswith("test"):
            total_test_rows += num_rows
            test_file_count += 1
    except Exception as e:
        print(f"Error reading {file}: {e}")

# Print individual file row counts
print("Row count per file:")
for name, count in file_row_counts.items():
    print(f"{name}: {count} rows")

# Print aggregate stats
print("\nAggregated row counts:")
print(f"Total train rows: {total_train_rows}")
print(f"Total val rows: {total_val_rows}")
print(f"Total test rows: {total_test_rows}")

# Print file counts
print("\nFile counts by type:")
print(f"Train files: {train_file_count}")
print(f"Val files:   {val_file_count}")
print(f"Test files:  {test_file_count}")

# Compute and print row ratios
total_rows = total_train_rows + total_val_rows + total_test_rows
print(f"")
if total_rows > 0:
    print("\nRow ratios (train : val : test):")
    ratio_train = total_train_rows / total_rows
    ratio_val = total_val_rows / total_rows
    ratio_test = total_test_rows / total_rows
    print(f"{ratio_train:.2f} : {ratio_val:.2f} : {ratio_test:.2f}")
else:
    print("No rows found to compute ratios.")

```

```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt

# Set folder path
folder_path = "/proj/sourasb-220503/TCSS-DIPA2/Fed-DIPA2/dataset_files/dynamic_FedDcprivacy"
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Storage for counts and verification
train_counts, val_counts, test_counts = [], [], []
train_files, val_files, test_files = {}, {}, {}

# Read and classify files
for file in csv_files:
    try:
        df = pd.read_csv(file)
        num_rows = len(df)
        file_name = os.path.basename(file)

        if file_name.startswith("train"):
            train_counts.append(num_rows)
            train_files[file_name] = num_rows
        elif file_name.startswith("val"):
            val_counts.append(num_rows)
            val_files[file_name] = num_rows
        elif file_name.startswith("test"):
            test_counts.append(num_rows)
            test_files[file_name] = num_rows
    except Exception as e:
        print(f"Error reading {file}: {e}")

# Print row counts for verification
print("\nTrain Files:")
for name, count in train_files.items():
    print(f"{name}: {count} rows")

print("\nValidation Files:")
for name, count in val_files.items():
    print(f"{name}: {count} rows")

print("\nTest Files:")
for name, count in test_files.items():
    print(f"{name}: {count} rows")

# Plotting function
def plot_distribution(sample_counts, title):
    if not sample_counts:
        print(f"[Warning] No data available to plot for {title}")
        return
    distribution = pd.Series(sample_counts).value_counts().sort_index()
    plt.figure(figsize=(10, 6))
    plt.bar(distribution.index, distribution.values)
    plt.xlabel("Number of Samples")
    plt.ylabel("Number of Annotators")
    plt.title(title)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

# Generate plots
plot_distribution(train_counts, "Train: Samples Distribution Across Annotators")
plot_distribution(val_counts, "Validation: Samples Distribution Across Annotators")
plot_distribution(test_counts, "Test: Samples Distribution Across Annotators")
```

```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt

# Parameters
folder_path = "/proj/sourasb-220503/TCSS-DIPA2/Fed-DIPA2/dataset_files/dynamic_FedDcprivacy"
threshold_delta = 0.5  # Change this to 0.6, 0.7 etc. if needed

# Get all CSV files
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Storage for train file statistics
train_files = {}  # {filename: num_rows}

# Collect only train files
for file in csv_files:
    file_name = os.path.basename(file)
    if file_name.startswith("train"):
        try:
            df = pd.read_csv(file)
            train_files[file_name] = len(df)
        except Exception as e:
            print(f"Error reading {file_name}: {e}")

# --- DR/DL User Identification Algorithm ---
total_samples = sum(train_files.values())

# Step 3: Compute data fractions
data_fractions = {f: c / total_samples for f, c in train_files.items()}

# Step 4: Sort by fraction descending
sorted_files = sorted(data_fractions.items(), key=lambda x: x[1], reverse=True)

# Step 510: Select top contributors until cumulative  threshold
U_DR = []
U_DL = []
cumulative = 0.0

for file_name, frac in sorted_files:
    if cumulative < threshold_delta:
        U_DR.append(file_name)
        cumulative += frac
    else:
        U_DL.append(file_name)

# Print results
print(f"\nTotal Train Files: {len(train_files)}")
print(f"U_DR (high contributors  {threshold_delta*100:.0f}% data): {len(U_DR)} files")
print(f"U_DL (low contributors): {len(U_DL)} files")

import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.lines import Line2D

def plot_dr_dl_split(train_files, U_DR, U_DL):
    values = [train_files[f] for f in U_DR + U_DL]
    colors = ['orange'] * len(U_DR) + ['steelblue'] * len(U_DL)

    plt.figure(figsize=(14, 6))
    bars = plt.bar(range(len(values)), values, color=colors)

    # Divider line after U_DR
    divider_position = len(U_DR) - 0.5
    plt.axvline(x=divider_position, color='black', linestyle='--', linewidth=2)

    # Custom legend with dotted line for 
    legend_elements = [
        Patch(facecolor='orange', label='U_DR ( 50% contributors)'),
        Patch(facecolor='steelblue', label='U_DL (< 50% contributors)'),
        Line2D([0], [0], color='black', linestyle='--', linewidth=2, label=' = 50% Divider')
    ]

    # Styling
    plt.xlabel("Annotators/Users (Sorted based on their contributions)", fontsize=24)
    plt.ylabel("Number of Training samples", fontsize=24)
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    plt.ylim(0, max(values) * 1.1)
    plt.legend(handles=legend_elements, loc='best', fontsize=20)
    plt.tight_layout()
    plt.show()


plot_dr_dl_split(train_files, U_DR, U_DL)
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.lines import Line2D

def plot_dr_dl_split(train_files, U_DR, U_DL):
    all_files_sorted = U_DR + U_DL
    values = [train_files[f] for f in all_files_sorted]
    colors = ['orange'] * len(U_DR) + ['steelblue'] * len(U_DL)

    # Identify clients with < 20 samples
    low_sample_positions = [i for i, f in enumerate(all_files_sorted) if train_files[f] < 20]
    low_sample_count = len(low_sample_positions)

    plt.figure(figsize=(14, 6))
    bars = plt.bar(range(len(values)), values, color=colors)

    # Divider for 50% threshold
    divider_position = len(U_DR) - 0.5
    plt.axvline(x=divider_position, color='black', linestyle='--', linewidth=2)

    # Add vertical dotted lines for clients with < 20 samples
    for pos in low_sample_positions:
        plt.axvline(x=pos, color='gray', linestyle=':', linewidth=1)

    # Legend
    legend_elements = [
        Patch(facecolor='orange', label='U_DR ( 50% contributors)'),
        Patch(facecolor='steelblue', label='U_DL (< 50% contributors)'),
        Line2D([0], [0], color='black', linestyle='--', linewidth=2, label=' = 50% Divider'),
        Line2D([0], [0], color='gray', linestyle=':', linewidth=2, label=f'Users with < 20 samples: {low_sample_count}')
    ]

    # Styling
    plt.xlabel("Annotators/Users (Sorted based on their contributions)", fontsize=24)
    plt.ylabel("Number of Training samples", fontsize=24)
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    plt.ylim(0, max(values) * 1.1)
    plt.legend(handles=legend_elements, loc='right', fontsize=24)
    plt.tight_layout()
    plt.show()

plot_dr_dl_split(train_files, U_DR, U_DL)
```
