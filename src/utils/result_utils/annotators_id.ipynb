```python
import os
import re
```

```python
folder_path = "/home/sourasb/PHD/Fed-DIPA2/clients"
files = os.listdir(folder_path)
extracted_from_strings = []
```

```python
pattern = r'^new_annotations_annotator(\d+)\.csv$'
```

```python
# Loop through files and extract numbers
for file in files:
    match = re.match(pattern, file)
    if match:
        extracted_from_strings.append(match.group(1))

# Print the extracted numbers
print(extracted_from_strings)
```

```python
identifier = 300
japan = []
uk = []

for i in range(len(extracted_from_strings)):
    if int(extracted_from_strings[i]) <= 300:
        japan.append(extracted_from_strings[i])
    else:
        uk.append(extracted_from_strings[i])
print(f"japan : {japan}")
print(f"uk : {uk}")
```
