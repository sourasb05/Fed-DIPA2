```python
import numpy as np
from scipy.stats import ttest_rel, wilcoxon
```

```python
feddc = [0.7735012355585843, 0.7703833054121043, 0.7689890972969733, 0.7836149069648293, 0.7472183667584597, 0.7223838646354701, 0.7595057921546593, 0.7301505347136024]

cfeddc = [0.7451223208578824, 0.765893910537889, 0.7349595007306904, 0.7499388585075953, 0.7456926157643319, 0.7670570032834444, 0.7424367328248866, 0.7313875465030607]
```

```python
# Paired t-test
t_stat, p_val = ttest_rel(feddc, cfeddc)
print("Paired t-test:", t_stat, p_val)

# Wilcoxon test
w_stat, p_val = wilcoxon(feddc, cfeddc)
print("Wilcoxon test:", w_stat, p_val)
```
